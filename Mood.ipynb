# Weekly Mood Journal

Welcome to my mood journal! As an experiment, I wanted to chronicle my moods in hour increments for one week to see how my moods fluctuated during the day. As a way to make the experiment "more challenging", I decided that I would do some basic coding so that my daily fluctuations in mood could be visualized via graphs.

For general understanding...there are graphs Tuesday through Saturday, with moods ranked from one to five (one for "I feel awful" and five for "this is the best I've ever felt!") and with the x-axis reflecting the time throughout the day.

When beginning this project I hypothesized that my moods would resemble a bell curve throughout the day (something along the lines of 1, 2, 3, 4, 5, 5, 4, 3, 2, 1) since I tend to wake up slowly, feel consistently pretty happy, and then lose energy as the day progresses; however, what the days seemed to reflect was there tend to be a lot of fluctuations in my mood throughout the day which I will address individually within each graph.

```python
import pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

def make_plot(filename):
    data = pandas.read_csv(filename, usecols=[0,1,2])
    
    fig, ax = plt.subplots(figsize=[20,10])

    for k, v in data.iterrows():
        ax.annotate(v["Notes"], (k, v["Mood"]),
                xytext=(0,0), textcoords='offset points',
               fontsize=15, color='black',
               rotation=45)


    data.plot(kind="line" , yticks=[1,2,3,4,5,6], x="Time", ax=ax)

```

```python
make_plot("mood_tuesday.csv")
```
